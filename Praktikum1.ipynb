# 3.2 Quantisierung
 In vielen Anwendungen, wie z.B. der digitale Verarbeitung analoger Signale, hat man es mit
wertkontinuierlichen Signalen zu tun. Die genaue Beschreibung der Amplitude dieser Signale
setzt eine unendliche Wortbreite voraus. Um solche Signale trotzdem verarbeiten zu
können, muss man sie zunächst auf eine endliche Anzahl von Amplitudenwerten quantisieren.
Dabei geht Information verloren, die auch später nicht mehr rekonstruiert werden kann.
Wählt man jedoch die Wortbreite ausreichend groß, so kann der hierbei gemachte Fehler
vernachlässigt werden. Er fällt der Sinke nicht auf oder kann toleriert werden (Irrelevanzreduktion).
Man unterscheidet zwischen skalaren Quantisierern und Vektorquantisierern. Skalare
Quantisierer bearbeiten jedes Quellensymbol einzeln, während Vektorquantisierer Blöcke
von Quellensymbolen zusammen verarbeiten. Bei den skalaren Quantisierern unterscheidet
man noch zwischen gleichförmigen (uniform) und ungleichförmigen (non-uniform)
Quantisierern. Wir wollen uns im folgenden nur mit skalaren Quantisierern beschäftigen.

# 3.3 Puls Code Modulation (PCM)
Bei PCM wird ein analoges Signal zunächst oberhalb der Nyquist Rate abgetastet, und die
Abtastwerte werden quantisiert. Dann werden die Abtastwerte mit m Bits codiert. Das verwendete
Codierschema ist dabei die natürliche binäre Codierung (NBC), bei der der niedrigste
Wert auf das Nullwort und der höchste Wert auf das All-Eins-Wort abgebildet wird.
Es gibt PCM mit gleichförmiger und ungleichförmiger (A-law, μ-law) Quantisierung.

## 3.3.1 Gleichförmige Quantisierung
Bei gleichförmiger Quantisierung wird das Intervall $[−x_{max} , x_{max}]$ der Größe $ 2x_{max} $ in N
gleiche Subintervalle der Größe $ \Delta=\frac{2x_{max}}{N} $ aufgeteilt.. Bei genügend großem N kann
man annehmen, dass die Wahrscheinlichkeitsverteilung in jedem Subintervall gleichförmig
ist. Die mittlere Leistung des Quantisierungsrauschens wird dann

### <center> $ D = \frac{\Delta^2}{12}=\frac{x_{max}^2}{3N^2} $ </center>

Wir nehmen im Folgenden an, dass $N$ eine Potenz von 2 ist: $N=2^v$
Für das Signal zu Quantisierungsgeräuschverhältnis SQNR ergibt sich mit : $\overline{X^2}$ der mittleren
Leistung des Nutzsignals:

### <center> $SQNR=3N^2\frac{\overline{X^2}}{x_{max}^2}$</center>

### <center> $SQNR_{dB}=4.8+6v+\frac{\overline{X^2}}{x_{max}^2}$</center>

<i>Anmerkung</i>: Für ein voll ausgesteuertes Sinussignal ist $\frac{\overline{X^2}}{x_{max}^2}=-3 dB$ für ein gleichverteiltes
Signal ergibt sich $\frac{\overline{X^2}}{x_{max}^2}=-4.8 dB$

## 3.3.2 Ungleichförmige Quantisierung

Bei der ungleichförmigen Quantisierung wird das Eingangssignal zunächst durch ein nichtlineares
Bauelement geschickt, um seinen dynamischen Bereich zu reduzieren. Der Ausgang
wird dann mit einem gleichförmigen Quantisierer verarbeitet.
Am Empfänger wird das Signal wieder durch die Inverse der Nichtlinearität korrigiert. Der
Gesamteffekt dieser Prozedur ist eine ungleichmäßige Quantisierung.
Bei der Übertragung von Sprachsignalen werden bei PCM entweder μ-law oder A-law
Nichtlinearitäten eingesetzt.
Die μ-law Nicht-Linearität ist durch die Beziehung

### <center>$y=g(x)=\frac{log(1+µ|x|)}{log(1+µ)}sgn(x)$</center>

gegeben,wobei $∣x∣≤1$ gelten muss und μ normalerweise auf den Wert 255 gesetzt wird.

#### M: Plotten Sie diese Charakteristik. Benutzen Sie dazu die Funktion mulaw().
Die Inverse der μ-law Charakteristik ergibt sich zu:
### <center>$x=\frac{(1+µ)^{|y|}-1}{µ}sgn(y)$</center>

```python
x=np.linspace(-1,1,201)
mu=255
y,a=mulaw(x,mu)
plt.plot(x,y)
plt.show()
```

# 4 Quellencodierung

Der Ausgang einer Informationsquelle (Daten, Video, Sprache etc.) kann als Zufallsprozess
modelliert werden. Der Informationsinhalt oder die Entropie eines diskreten, gedächtnislosen
und stationären Zufallsprozesses wird definiert als:
### <center>$H(X)=-\sum\limits_{\substack{x\in X}} p(x)ld( p(x))$</center>
und gibt die mittlere Anzahl von Bits pro Quellensymbol an,(die notwendig ist, um die Quelle
eineindeutig zu repräsentieren. Die mittlere Anzahl von Bits pro Quellensymbol eines
realen Quellencodes kann zwar der Entropie beliebig nah kommen, sie aber niemals unterschreiten.
Für eine Quelle mit nur zwei Zeichen (binärer Fall) ergibt sich für die Entropie:
### <center>$H_b(p)=-pld(p)-(1-p)ld(1-p)$</center>

#### F: Skizzieren Sie die Entropie für den binären Fall. Welche Auftrittswahrscheinlichkeit $p$ für die beiden Zeichen maximiert diese Entropie? (Tip: Ableitung der Funktion $H_b(p)$ nach $p$ und gleich Null setzen!)

>Es gibt verschiedene Algorithmen zur rauschfreien Quellencodierung (noiseless coding).
Die bekanntesten sind die Huffman-Codierung , der Lempel-Ziv Algorithmus und die beim
Fax angewandte Lauflängencodierung.
Wir wollen uns im folgenden mit der Huffman Codierung als Beispiel für eine präfixfreie Codierung
befassen.
